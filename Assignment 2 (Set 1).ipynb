```python
Q1=np.quantile(measure_x,0.25)
Q3=np.quantile(measure_x,0.75)
med = np.median(measure_x)
IQR= Q3- Q1
upper_bound = Q3+(1.5*IQR)
lower_bound = Q1-(1.5*IQR)
```

```python
print('\n First Quantile :',Q1,
     '\n Second Quantile :',med,
     '\n Third Quantile :',Q3,
     '\n Inter-Quantile Range :',IQR,
     '\n Upper Whisker :',upper_bound,
     '\n Lower Whisker :',lower_bound)
```

```python
Outliers = measure_x[(measure_x<=lower_bound)|(measure_x>=upper_bound)]
```
